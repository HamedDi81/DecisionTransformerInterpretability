```python
from transformer_lens import HookedTransformer


# get gpt2-small from pretrained
model = HookedTransformer.from_pretrained(
    "gpt2-small",
    center_unembed=True,
    center_writing_weights=True,
    fold_ln=True,
    refactor_factored_attn_matrices=True,
)
```

```python
# get the tokenizer vocabulary
tokenizer = model.tokenizer
vocabulary = tokenizer.get_vocab()
vocabulary
```

```python
vocabulary.keys()
```

```python
word = 'ĠDiablo'
for i in range(len(word)):
    print(word[:i] + 'a' + word[i+1:])
```

```python
#

from collections import defaultdict

def get_a_letter_diff_pairs(words):
    pairs = list()
    word_set = set(words)  # Convert list to set for O(1) lookups

    for word in words:
        for i in range(len(word)):
            a_word = word[:i] + 'i' + word[i+1:]
            # Only add to the a_word list if it's in the original word list
            # substituted letter is 

            if a_word in word_set and word[i] == 'e':
                pairs.append((a_word, word))

    return pairs

pairs = get_a_letter_diff_pairs(vocabulary.keys())
print(len(pairs))
print(pairs[:10])
```

```python
# get the first items as a list
first_items = [pair[0] for pair in pairs]
# get the token ids for the first items


# get the second items as a list
second_items = [pair[1] for pair in pairs]
# get the token ids for the second items

# what are the distribution of lengths in first_items 
# and second_items
first_items_lengths = [len(item) for item in first_items]
second_items_lengths = [len(item) for item in second_items]

# filter by lengths == 4 
first_items = [item for item in first_items if len(item) == 4]
second_items = [item for item in second_items if len(item) == 4]

# filter for the two words being the same 
mask = [first_items[i] != second_items[i] for i in range(len(first_items))]
first_items = [first_items[i] for i in range(len(first_items)) if mask[i]]
second_items = [second_items[i] for i in range(len(second_items)) if mask[i]]

# filter for only lowercased words
mask = [first_items[i].islower() and second_items[i].islower() for i in range(len(first_items))]
first_items = [first_items[i] for i in range(len(first_items)) if mask[i]]
second_items = [second_items[i] for i in range(len(second_items)) if mask[i]]

first_items_token_ids = [vocabulary[item] for item in first_items]
second_items_token_ids = [vocabulary[item] for item in second_items]

print(len(first_items))
print(first_items[:10])
print(second_items[:10])
```

```python
model.W_E
```

```python
# get the token embeddings for the first items
first_items_token_embeddings = model.W_E[first_items_token_ids]
first_items_token_embeddings.shape

# get the token embeddings for the second items
second_items_token_embeddings = model.W_E[second_items_token_ids]
second_items_token_embeddings.shape

# take the difference between the first and second items
diff = first_items_token_embeddings - second_items_token_embeddings
diff.shape
```

```python
# get the norms of the differences
diff_norms = diff.norm(dim=-1).detach()
diff_norms.shape

# plot the distribution of the norms
import plotly.express as px
fig = px.histogram(diff_norms, nbins=100)
fig.show()
```

```python
# divide each difference by its norm
normalized_diff = diff.div(diff_norms.unsqueeze(-1) + 1e-8)
print(normalized_diff.max())

# randomly sample 500 differences
import torch
indices = torch.randperm(normalized_diff.shape[0])[:500]
sampled_norms = normalized_diff[indices]
labels = [f"{first_items[i]}-{second_items[i]}" for i in indices]
print(sampled_norms.shape)

# get cosine similarity between the sampled differences
cosine_similarity = sampled_norms @ sampled_norms.T
cosine_similarity = cosine_similarity.detach()
cosine_similarity.shape
```

```python
import pandas as pd 
import numpy as np 
from scipy.cluster import hierarchy

df = pd.DataFrame(cosine_similarity.detach().numpy())
df.columns = labels
df.index = labels

data_array = df.to_numpy()
linkage = hierarchy.linkage(data_array)
dendrogram = hierarchy.dendrogram(
    linkage, no_plot=True, color_threshold=-np.inf
)

reordered_ind = dendrogram["leaves"]
# reorder df by ind
df = df.iloc[reordered_ind, reordered_ind]
# data_array = df.to_numpy()

fig = px.imshow(
    df,
    color_continuous_midpoint=0,
    color_continuous_scale='RdBu',
    labels=dict(color='Cosine Similarity'),
    width=800,
    height=800,
)

# don't label either axis ticks
fig.update_xaxes(showticklabels=False)
fig.update_yaxes(showticklabels=False)

fig.show()
```

```python
# search for any rows/collumns whose middle two letter are the 'll'
mask = df.index.str.contains('ll') & ~df.index.str.startswith('l') & ~df.index.str.endswith('l')
# mask both rows and collumns
tmp = df.loc[mask, mask]

fig = px.imshow(
    tmp,
    color_continuous_midpoint=0,
    color_continuous_scale='RdBu',
    labels=dict(color='Cosine Similarity'),
    width=800,
    height=800,
)

# don't label either axis ticks
fig.update_xaxes(showticklabels=False)
fig.update_yaxes(showticklabels=False)

fig.show()
```
